```
# =========================================================
# === 🐍 Comprehensive Python Crash Course (Interview Prep) 🐍 ===
# =========================================================

# ========================================================
# ==================== 1. Basic Python ===================
# ========================================================
# This section covers the fundamental building blocks of the Python language.
# Mastering these basics is the first step to becoming a proficient developer.

# --- 1.1 Variables and Data Types ---
#
# Lecture:
# Think of a **variable** as a label or a named box that holds a piece of data.
# In Python, you don't need to specify what kind of data the box will hold.
# The language is **dynamically typed**, which means Python automatically
# figures out the data type for you. For example, if you put the number `10`
# in a box named `age`, Python knows `age` is an integer. If you put the word
# `"Alice"` in a box named `name`, Python knows `name` is a string.
#
# A key concept for data types is **mutability**. This simply means whether the
# data can be changed after it's created. If it's **mutable**, you can change it.
# If it's **immutable**, you cannot.
#
# The core built-in data types are:
# 1. **String (`str`)**: Used for text. It is an ordered sequence of characters
#    (letters, numbers, symbols) enclosed in quotes. Strings are **immutable**.
#    If you try to "change" a string, Python actually creates a brand new string
#    in memory, which is an important concept for memory management.
# 2. **Integer (`int`)**: Used for whole numbers (e.g., -5, 0, 100). They are **immutable**.
# 3. **Float (`float`)**: Used for numbers with a decimal point (e.g., 3.14, -0.5).
#    They are also **immutable**.
# 4. **Boolean (`bool`)**: Used for truth values: `True` or `False`. These are
#    fundamental for making decisions in your code.
# 5. **NoneType (`None`)**: Represents the absence of a value. It's like an
#    empty box with no data in it, often used as a placeholder.
#
# You can change a variable's data type using **type conversion** (also called
# casting) with functions like `int()`, `float()`, `str()`, and `bool()`.

# String (str): An immutable sequence of characters.
my_string = "Hello, Python!"
print(f"1.1 String: {my_string}")
# Concatenation creates a NEW string in memory.
new_string = my_string + " It's a great day."
print(f"1.1 New string after concatenation: {new_string}")

# Integer (int) and Float (float): Immutable numeric types.
my_integer = 100
my_float = 99.99
print(f"1.1 Integer: {my_integer}, Float: {my_float}")

# Boolean (bool): Used for logical operations.
is_learning = True
print(f"1.1 Boolean (is_learning): {is_learning}")

# NoneType: Represents no value.
my_none = None
print(f"1.1 NoneType variable: {my_none}")

# Type Conversion (Casting)
string_number = "500"
converted_integer = int(string_number)
print(f"1.1 String to Integer: {converted_integer}, Type: {type(converted_integer)}")
print("-" * 35)


# --- 1.2 Basic Data Structures ---
#
# Lecture:
# Data structures are ways to organize and store collections of data. Knowing
# which one to use is crucial. Let's compare the four basic types:
#
# | Data Structure | Mutability  | Ordered? | Duplicates? | Description & Best Use Case                                |
# |----------------|-------------|----------|-------------|------------------------------------------------------------|
# | **List** | Mutable     | Yes      | Yes         | A flexible, ordered collection. Use for storing items that will change. |
# | **Tuple** | Immutable   | Yes      | Yes         | A fixed, ordered collection. Use for data that should not change. |
# | **Set** | Mutable     | No       | No          | An unordered collection of unique items. Great for removing duplicates. |
# | **Dictionary** | Mutable     | Yes (since 3.7) | No for keys | A collection of key-value pairs. Use for fast lookups by name (key). |
#
# - **Lists** are the most common. You can add, remove, and change items after
#   the list is created.
# - **Tuples** are like lists, but they are "locked." You can't change their
#   contents. This makes them faster and safer for data that shouldn't be altered,
#   like the coordinates of a point.
# - **Sets** are like a club where every member must be unique. They are useful
#   when you only care about the existence of an item, not its order.
# - **Dictionaries** are like a real-world dictionary or phone book. You look
#   up a value (like a phone number) using a unique key (like a person's name).
#   The keys must be immutable (like strings or numbers) so they can be looked up quickly.

# List: Ordered, mutable, allows duplicates.
my_list = ["apple", "banana", "cherry"]
print(f"1.2 Initial List: {my_list}")
my_list.append("date")         # Add an item to the end.
my_list.insert(1, "avocado")   # Insert at a specific index.
my_list[0] = "blueberry"       # Modify an item.
print(f"1.2 List after modifications: {my_list}")
del my_list[2]                 # Remove an item by index.
print(f"1.2 List after deleting 'cherry': {my_list}")

# Tuple: Ordered, immutable, allows duplicates.
my_tuple = (10, 20, 30)
print(f"1.2 Tuple: {my_tuple}")
# The following line would raise a TypeError:
# my_tuple[0] = 5

# Set: Unordered, mutable, unique elements.
my_set = {1, 2, 2, 3, 4, 4}
print(f"1.2 Set (duplicates removed): {my_set}")
my_set.add(5)
print(f"1.2 Set after adding 5: {my_set}")
is_present = 3 in my_set
print(f"1.2 Is 3 in the set? {is_present}")

# Dictionary: Key-value pairs.
my_dict = {"name": "Charlie", "age": 25, "city": "Paris"}
print(f"1.2 Dictionary: {my_dict}")
my_dict["email"] = "charlie@example.com"
print(f"1.2 Value for key 'name': {my_dict['name']}")
print(f"1.2 Keys: {list(my_dict.keys())}, Values: {list(my_dict.values())}")
# Using .get() is safer for accessing keys that might not exist.
print(f"1.2 Getting non-existent key with .get(): {my_dict.get('country', 'Unknown')}")
print("-" * 35)


# --- 1.3 Operators and Expressions ---
#
# Lecture:
# Operators are symbols that tell Python to perform a certain action on values.
# An **expression** is a combination of values and operators that results in a
# single value. For example, `10 + 5` is an expression that evaluates to `15`.
#
# There are a few main types of operators:
# - **Arithmetic Operators**: For math operations. `+` (addition), `-` (subtraction),
#   `*` (multiplication), `/` (division), `//` (floor division, which gives you
#   the whole number part of a division), `%` (modulus, which gives the remainder),
#   and `**` (exponentiation, for powers like `2**3` which is 8).
# - **Assignment Operators**: Used to assign a value to a variable. The `=` is
#   the basic assignment operator, but you also have shortcuts like `+=`, which
#   is a combination of addition and assignment. For example, `x += 5` is the
#   same as `x = x + 5`.
# - **Comparison Operators**: Used to compare two values. They always result in
#   either `True` or `False`. These include `==` (equal to), `!=` (not equal to),
#   `>` (greater than), `<` (less than), and so on.
# - **Logical Operators**: Used to combine conditional statements. These are `and`,
#   `or`, and `not`.
#
# **Operator precedence** is like the "order of operations" from math class.
# It tells Python which operators to evaluate first. You can always use parentheses
# `()` to control the order.

# Assignment operators
x = 10
x += 5  # Equivalent to x = x + 5
print(f"1.3 x after += 5: {x}")

# Comparison and logical operators
age = 18
has_id = True
# The expression (age >= 18) and (has_id) evaluates to True.
if age >= 18 and has_id:
    print("1.3 Allowed to enter.")
else:
    print("1.3 Not allowed to enter.")
print("-" * 35)


# --- 1.4 Control Flow: Conditionals and Loops ---
#
# Lecture:
# **Control flow** is about the order in which your code runs. By default, it
# runs from top to bottom. But with conditionals and loops, you can change that.
#
# - **Conditionals (`if`/`elif`/`else`)**: These are your code's decision-making
#   tools. The `if` statement checks if a condition is `True`. If it is, the
#   code inside the `if` block runs. If not, Python moves to the `elif` (else if)
#   statements, checking them in order. If none of the `if` or `elif` conditions
#   are `True`, the code in the final `else` block runs.
#
# - **`for` loop**: This is used when you know exactly how many times you want to
#   repeat an action. It's great for iterating over all the items in a collection
#   like a list or tuple.
#
# - **`while` loop**: This is used when you want to repeat an action as long as
#   a certain condition remains `True`. **Warning:** Be careful to make sure the
#   condition will eventually become `False`, or your program will run forever
#   in an **infinite loop**.
#
# You can also use special keywords to control loops:
# - **`break`**: Immediately stops the loop and exits it completely.
# - **`continue`**: Skips the rest of the current turn of the loop and moves
#   on to the next one.

# Conditionals
temperature = 25
if temperature > 30:
    print("1.4 It's a hot day.")
elif temperature > 20:
    print("1.4 It's a pleasant day.")
else:
    print("1.4 It's a cool day.")

# For loops with a list
print("\n1.4 For Loop: Iterating through a list:")
for fruit in my_list:
    if fruit == "banana":
        continue # Skip printing "banana"
    print(f"  - I have a {fruit}")

# While loops: indefinite iteration.
print("\n1.4 While Loop:")
counter = 3
while counter > 0:
    print(f"  - {counter}...")
    counter -= 1
    if counter == 1:
        break # Exit the loop early
print("-" * 35)


# --- 1.5 Functions and Variable Scope ---
#
# Lecture:
# Functions are named blocks of code that perform a specific task. They are
# incredibly useful for two main reasons:
# 1. **Reusability**: You can write a piece of code once and then use it many
#    times without rewriting it. This is the **DRY (Don't Repeat Yourself)** principle.
# 2. **Organization**: They make your code cleaner, easier to read, and more
#    manageable by breaking it down into smaller, logical chunks.
#
# - **Parameters vs. Arguments**: A **parameter** is a variable you list in the
#   function's definition, like a placeholder. An **argument** is the actual
#   value you pass into the function when you call it.
# - **`return` statement**: This is how a function sends a value back to the
#   place where it was called. If a function doesn't have a `return` statement,
#   it automatically returns a special value: `None`.
# - **Scope and Global Variables**: The **scope** of a variable refers to where
#   it can be accessed. Variables created inside a function have **local scope**
#   and can only be used within that function. Variables created outside of any
#   function have **global scope** and can be accessed from anywhere.
#
#   Normally, a function can read a global variable, but it cannot change it.
#   If you want a function to modify a global variable, you must use the `global`
#   keyword inside the function before you use the variable's name. This tells
#   Python: "Don't create a new local variable with this name; I want to modify
#   the existing global one."

# Global variable
global_message = "I am a global message."

def say_hello(name="Guest"):
    # This function can READ the global_message
    print(f"1.5 Hello, {name}! ({global_message})")

def change_global_message():
    # To MODIFY the global variable, we must use the `global` keyword.
    global global_message
    global_message = "I have been changed by a v."
    print(f"1.5 Inside the function, the global message is now: {global_message}")

say_hello("Alice") # Reads the original global message
change_global_message() # Modifies the global message
say_hello("Bob") # Reads the new, modified global message
print("-" * 35)


# ========================================================
# ================== 2. Intermediate Python ==============
# ========================================================

# --- 2.1 List Comprehensions and Generators ---
#
# Lecture:
# - **List Comprehensions**: This is a powerful, concise way to create a list.
#   Instead of writing a multi-line `for` loop to build a list, you can do it
#   in a single, readable line. It's a "shortcut" for creating lists based on an
#   existing list or other collection. It builds the entire list in memory all at once.
#   The basic structure is `[expression for item in iterable if condition]`.
#
# - **Generators**: Generators are like an "on-demand" version of a list comprehension.
#   They don't build the entire list in memory. Instead, they create and `yield`
#   (produce) one item at a time only when you ask for it. This is called
#   **lazy evaluation**. Generators are extremely useful when you're working with
#   very large datasets, as they can save a huge amount of memory. They use
#   parentheses `()` instead of the square brackets `[]` of a list comprehension.

# List comprehension:
squares_comprehension = [i**2 for i in range(10) if i % 2 == 0]
print(f"2.1 List comprehension (even numbers squared): {squares_comprehension}")

# Generator expression:
# Note the use of parentheses instead of brackets.
# This does not create a list of a billion items, just a way to generate them.
large_generator = (i**2 for i in range(1000000000))
print(f"\n2.1 Using a generator for memory efficiency. Next item: {next(large_generator)}")
print("-" * 35)


# --- 2.2 Functions: Arguments, Lambdas, and Decorators ---
#
# Lecture:
# - **`*args` and `**kwargs`**: These are special syntaxes that allow a function
#   to accept a **variable number of arguments**. This makes your function more
#   flexible because you don't need to know in advance how many arguments the
#   user will pass.
#
#   **`*args` for Positional Arguments**:
#   - The asterisk `*` gathers all the extra **positional arguments** you pass
#     to a function into a single variable.
#   - This variable is a **tuple**.
#   - The name `args` is a convention, but you can use any name after the `*`
#     (e.g., `*numbers`).
#   - Use `*args` when you want a function to be able to accept any number of
#     arguments that don't have a specific keyword.
#
#   **`**kwargs` for Keyword Arguments**:
#   - The double asterisk `**` gathers all the extra **keyword arguments**
#     (arguments passed with a `key=value` syntax) into a single variable.
#   - This variable is a **dictionary**.
#   - The name `kwargs` is a convention, but you can use any name after the `**`
#     (e.g., `**details`).
#   - Use `**kwargs` when you want a function to handle any number of named
#     arguments.
#
#   You can use both `*args` and `**kwargs` in the same function definition.
#   The order is always `def my_func(some_param, *args, **kwargs):`.
#
#   **Lambda Functions**: These are small, anonymous, one-line functions
#   defined with the `lambda` keyword. They are restricted to a single expression
#   and are typically used as arguments for higher-order functions like
#   `map()`, `filter()`, or `sorted()`.
#
#   **Decorators**: A design pattern that allows you to **modify or extend
#   the behavior of a function without permanently changing its source code**.
#   A decorator is a function that takes another function as an argument,
#   adds some functionality, and returns a new function (a "wrapper"). The
#   `@` syntax is syntactic sugar for this process.

def calculate_sum(*numbers):
    """*numbers collects positional arguments into a tuple."""
    print(f"2.2 `*numbers` received: {numbers}, Type: {type(numbers)}")
    return sum(numbers)

# This function can be called with a varying number of arguments
print(f"2.2 Sum of three numbers: {calculate_sum(90, 85, 78)}")
print(f"2.2 Sum of two numbers: {calculate_sum(95, 80)}")

def display_profile(**details):
    """**details collects keyword arguments into a dictionary."""
    print("\n2.2 User Profile:")
    print(f"2.2 `**details` received: {details}, Type: {type(details)}")
    for key, value in details.items():
        print(f"  - {key}: {value}")

# This function can be called with a varying number of keyword arguments
display_profile(name="Eve", profession="Data Scientist", company="TechCorp")
display_profile(name="Bob", age=30)

# Lambda function example
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x**2, numbers))
print(f"\n2.2 Using a lambda function with map: {squared_numbers}")

# Decorator example
def logger_decorator(func):
    """A decorator that logs a function's call."""
    import datetime
    def wrapper(*args, **kwargs):
        timestamp = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        print(f"2.2 LOG: Function '{func.__name__}' was called at {timestamp}.")
        return func(*args, **kwargs)
    return wrapper

@logger_decorator
def greet(name):
    print(f"Hello, {name}!")

print("\n2.2 Demonstrating a decorator:")
greet("Dave")
print("-" * 35)


# --- 2.3 Object-Oriented Programming (OOP) ---
#
# Lecture:
# **Object-Oriented Programming (OOP)** is a way of organizing your code around
# **objects**. Think of an object as a digital version of a real-world thing,
# like a car or a dog. An object has two main parts:
# 1. **Attributes**: The data or properties of the object (e.g., a dog's `name` and `breed`).
# 2. **Methods**: The actions or behaviors the object can perform (e.g., a dog can `speak`).
#
# We use a **class** as a blueprint to create these objects. The main principles of OOP are:
# - **Encapsulation**: This means bundling the data (attributes) and the
#   methods that work on that data into a single unit (the class). It's like a
#   car's engine—you can use it without needing to know all the inner workings.
# - **Inheritance**: This allows a new class (a "child" class) to inherit
#   the attributes and methods of an existing class (a "parent" class). This
#   is great for code reuse. For example, a `Dog` class can inherit from an
#   `Animal` class and get all of its basic properties automatically.
# - **Polymorphism**: This means "many forms." It's the ability of different
#   objects to respond to the same method call in their own unique way. For
#   example, if you call a `speak()` method on a `Dog` object, it barks.
#   If you call it on a `Cat` object, it meows.
#
# - **`__init__`**: This is a special method called a **constructor**. It's
#   automatically run when you create a new object from a class. Its job is
#   to set up the initial state of the object, like giving a new dog its name.
# - **`self`**: This refers to the specific object that is currently using the
#   method. It's always the first parameter of any method in a class.

class Animal:
    """A blueprint for creating animal objects."""
    def __init__(self, name, species):
        self.name = name
        self.species = species

    def speak(self):
        """A polymorphic method to be overridden by subclasses."""
        print(f"2.3 {self.name} makes a sound.")

class Dog(Animal):
    """A subclass inheriting from Animal."""
    def __init__(self, name, breed):
        # Call the parent class's constructor to initialize inherited attributes.
        super().__init__(name, species="Dog")
        self.breed = breed

    # Method overriding: Dog provides its own implementation of 'speak'.
    def speak(self):
        print(f"2.3 {self.name} the {self.breed} barks! Woof woof!")

# Method types:
class MyClass:
    class_variable = "I'm a class variable."

    def my_instance_method(self):
        print(f"2.3 This is an instance method. I can access 'self'.")

    @classmethod
    def my_class_method(cls):
        print(f"2.3 This is a class method. I can access 'cls' and class variables: {cls.class_variable}")

    @staticmethod
    def my_static_method():
        print("2.3 This is a static method. I have no access to 'self' or 'cls'.")

my_dog = Dog("Buddy", "Golden Retriever")
my_dog.speak()

instance = MyClass()
instance.my_instance_method()
MyClass.my_class_method()
MyClass.my_static_method()
print("-" * 35)


# --- 2.4 Error and Exception Handling ---
#
# Lecture:
# Errors happen. An **exception** is an error that occurs while your program is
# running. Good programmers write code that can handle these exceptions gracefully,
# so the program doesn't crash. We use a "safety net" with `try` and `except` blocks.
#
# - **`try`**: You put the code that might cause an error inside this block.
#   If an error occurs, Python stops running the `try` block and jumps to the
#   `except` block.
# - **`except`**: This block is like a catcher's mitt. It catches a specific type
#   of error. You can have multiple `except` blocks to handle different errors
#   in different ways. For example, you might handle a `ValueError` (invalid
#   data) differently than a `TypeError` (wrong data type).
# - **`else`**: This optional block only runs if the code inside the `try` block
#   completes **without any errors**.
# - **`finally`**: This optional block **always runs**, no matter what. It's
#   perfect for cleanup tasks, like closing a file, that must happen even if
#   an error occurred.

try:
    # This code will raise a ValueError because 'hello' is not a number.
    number = int("hello")
except ValueError as e:
    # The 'as e' part assigns the exception object to a variable 'e'
    print(f"2.4 Error: Invalid input, expected a number. Details: {e}")
except TypeError:
    print("2.4 Error: A type-related issue occurred.")
else:
    # This block is skipped because a ValueError was raised
    print(f"2.4 The number is: {number}")
finally:
    print("2.4 Exception handling block finished. This always runs.")
print("-" * 35)

# --- 2.5 Type Hinting for Better Code Clarity ---
#
# Lecture:
# **Type hinting** is a modern feature in Python that helps you make your code
# more readable and robust. It's a way of suggesting what data type a variable
# or function argument should be, or what data type a function should return.
# This doesn't change how your code runs, as Python is still dynamically typed,
# but it's a valuable tool for a few reasons:
#
# 1. **Clarity**: It makes your code easier for other programmers (and your future
#    self!) to understand. You immediately know what kind of data is expected.
# 2. **Tooling**: Code editors and static analysis tools can use these hints to
#    find potential bugs or type mismatches before you even run your code.
# 3. **Documentation**: It serves as a form of self-documentation, explaining
#    the purpose of your code more clearly.
#
# The syntax for type hinting a function return value is `def function_name() -> type:`.
# We'll use the example of a `DataFrame`, a popular data structure from the
# pandas library, often used for data science and analysis.
#
# Note: You need to import the `DataFrame` type to use it in a type hint.

import pandas as pd
from pandas import DataFrame

def create_sample_data() -> DataFrame:
    """This function is *hinted* to return a pandas DataFrame."""
    data = {
        "Name": ["Alice", "Bob", "Charlie"],
        "Age": [25, 30, 35]
    }
    return pd.DataFrame(data)

# When you call the function, you know it should give you a DataFrame.
my_dataframe = create_sample_data()
print("2.5 Type Hinting:")
print(my_dataframe)
print(f"The returned object is of type: {type(my_dataframe)}")
print("-" * 35)


# ========================================================
# =================== 3. Advanced Python =================
# ========================================================

# --- 3.1 Concurrency with Multithreading ---
#
# Lecture:
# **Concurrency** is about a program managing multiple tasks at the same time,
# giving the appearance of running in parallel, even on a single processor.
#
# - **Multithreading** is one way to achieve this. A **thread** is a lightweight
#   process. When you use multithreading, you split your program's work into
#   multiple threads that the operating system can switch between. This is
#   most effective for **I/O-bound tasks**, which are tasks that spend most of
#   their time waiting for something to happen, like a file to load or a
#   network request to return. While one thread is waiting, the OS can let
#   another thread do its work.
#
# - A key concept to understand is the **Global Interpreter Lock (GIL)**.
#   In Python (specifically CPython), the GIL prevents more than one thread
#   from executing Python code at the exact same time. This means multithreading
#   is not very useful for **CPU-bound tasks** (heavy calculations), as they
#   all have to wait for the same lock.

import threading, time

def slow_task(name):
    print(f"3.1 Thread '{name}': Starting...")
    time.sleep(2) # Simulates an I/O wait, like downloading a file
    print(f"3.1 Thread '{name}': Finishing.")

thread1 = threading.Thread(target=slow_task, args=("Download",))
thread2 = threading.Thread(target=slow_task, args=("Process",))

thread1.start()
thread2.start()

thread1.join() # Wait for thread1 to complete before moving on
thread2.join() # Wait for thread2 to complete before moving on
print("-" * 35)


# --- 3.2 Asynchronous Programming with Asyncio ---
#
# Lecture:
# **Asynchronous programming** is another way to handle multiple tasks concurrently,
# but it works differently than multithreading. It uses a **single thread** and
# an **event loop**.
#
# Imagine a chef working in a kitchen. The chef can start a pot of water boiling
# (`async`), then turn to chop vegetables while the water heats up. When the water
# is boiling, the chef gets a signal to go back to that task. This is the core
# idea of **asyncio**. When one task needs to wait for an I/O operation (like
# waiting for a webpage to load), it can pause itself and let the main program
# do other work.
#
# - **`async` and `await`**:
#   - An `async def` function defines a **coroutine**. This tells Python that
#     this function is a task that can be paused and resumed.
#   - The `await` keyword is used inside an `async` function to say: "I'm going
#     to wait here for a while. You can go do other work now and come back to
#     me when I'm ready."
#
# `asyncio` is highly efficient for **I/O-bound tasks** because it avoids the
# overhead of multiple threads, making it a great choice for web applications
# and network-heavy code.

import asyncio

async def fetch_data(url):
    print(f"3.2 Async task: Fetching {url}...")
    await asyncio.sleep(2) # Pauses this coroutine, letting other tasks run
    print(f"3.2 Async task: Finished {url}.")

async def main_async():
    # `asyncio.gather` runs coroutines concurrently
    await asyncio.gather(fetch_data("api1"), fetch_data("api2"))

asyncio.run(main_async())
print("-" * 35)


# --- 3.3 Modules, Packages, and `if __name__ == '__main__':` ---
#
# Lecture:
# As your programs grow, you'll want to organize your code to keep it manageable
# and reusable. That's where modules and packages come in.
#
# - A **module** is simply a single Python file (`.py`). It's like a single
#   toolbox filled with related tools (functions, classes, variables). You can
#   use the `import` statement to bring these tools into your current file.
#
# - A **package** is a collection of modules organized in a directory. Think of a
#   package as a large toolbox that contains many smaller, specialized toolboxes.
#   A directory is treated as a package if it contains an `__init__.py` file
#   (though this is optional in modern Python 3.3+).
#
# - **`if __name__ == '__main__':`**: This is a crucial line you'll see in many
#   Python scripts. When you run a Python file directly from the command line,
#   Python assigns the special variable `__name__` the value `'__main__'`.
#   However, if you `import` that same file as a module into another script,
#   Python assigns `__name__` the name of the module instead.
#
#   This block of code lets you define code that should **only run when the script
#   is executed directly**, not when it's imported. It's a best practice for
#   keeping your code reusable.

import os
import sys
from collections import Counter

print(f"3.3 Current working directory: {os.getcwd()}")
print(f"3.3 Python version: {sys.version.split()[0]}")
# Example of using a function from an imported module
count = Counter("banana")
print(f"3.3 Character count from collections module: {count}")

def main():
    """This function holds the main logic for the script."""
    print("\n3.3 This is a test message from the `main` function.")

# This code block will only execute if this script is run directly.
if __name__ == "__main__":
    print("\n3.3 The `if __name__ == '__main__':` block is running.")
    main()

print("-" * 35)
```
